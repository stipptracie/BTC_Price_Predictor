```python
from pandas.tseries.offsets import DateOffset
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import hvplot.pandas
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
```

```python

df = pd.read_csv('BTC.csv')
```

```python
df.set_index('Date',inplace=True)
```

```python
df.index = pd.to_datetime(df.index)
df
```

```python
df['Pct Change'] = df['Close'].pct_change()
df
```

```python
#SVM MODEL
df['Signal'] = 0
df.loc[(df['Pct Change'] >=0), 'Signal'] = 1
df.loc[(df['Pct Change'] <0), 'Signal'] = 0
```

```python
X = df[['Close', 'Volume', 'Pct Change', '50 SMA', '200 SMA', 'RSI']].shift().dropna()
y = df['Signal'].dropna()
```

```python
df.tail()
```

```python
training_begin = X.index.min()
training_end = X.index.min() + DateOffset(months=12)
```

```python
X_train = X.loc[training_begin:training_end]
y_train = y.loc[training_begin:training_end]
```

```python
X_test = X.loc[training_end+DateOffset(days=1):]
y_test = y.loc[training_end+DateOffset(days=1):]
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
svm_model = SVC()
svm_model = svm_model.fit(X_train_scaled, y_train)
svm_pred = svm_model.predict(X_test_scaled)
svm_pred
```

```python
svm_testing_report = classification_report(y_test, svm_pred)
print(svm_testing_report)
```

```python
# Create a new empty predictions DataFrame.

# Create a predictions DataFrame
predictions_df =  pd.DataFrame(index=X_test.index)

# Add the SVM model predictions to the DataFrame
predictions_df['Predicted'] = svm_pred

# Add the actual returns to the DataFrame
predictions_df['Pct Change'] = df["Pct Change"]

# Add the strategy returns to the DataFrame
predictions_df['Strategy Returns'] =  (
    df["Pct Change"] * predictions_df["Predicted"]
)

# Review the DataFrame
display(predictions_df.head())
display(predictions_df.tail())
```

```python
# Plot the actual returns versus the strategy returns
long_period_plot = (1 + predictions_df[['Pct Change', 'Strategy Returns']]).cumprod().hvplot(title=' long sma vs Cummulative Returns',
                                                                                            ylabel='%')

long_period_plot
```

```python
#logistic regression
logistic_regression_model = LogisticRegression(random_state=7)
lr_model =logistic_regression_model.fit(X_train_scaled, y_train)
lr_pred = lr_model.predict(X_test_scaled)
lr_pred
```

```python
lr_testing_report = classification_report(y_test, lr_pred)
print(lr_testing_report)
```

```python
# Create a new empty predictions DataFrame.

# Create a predictions DataFrame
lrpredictions_df =  pd.DataFrame(index=X_test.index)

# Add the SVM model predictions to the DataFrame
lrpredictions_df['Predicted'] = lr_pred

# Add the actual returns to the DataFrame
lrpredictions_df['Pct Change'] = df["Pct Change"]

# Add the strategy returns to the DataFrame
lrpredictions_df['Strategy Returns'] =  (
    df["Pct Change"] * lrpredictions_df["Predicted"]
)

# Review the DataFrame
display(lrpredictions_df.head())
display(lrpredictions_df.tail())
```

```python
# Plot the actual returns versus the strategy returns
long_period_plot = (1 + lrpredictions_df[['Pct Change', 'Strategy Returns']]).cumprod().hvplot(title=' ',
                                                                                            ylabel='%')

long_period_plot
```

```python
#XGB Classifier
XGB = XGBClassifier()
XGB.fit(X_train, y_train)
XGB = XGB.predict(X_test_scaled)
print(XGB)

XGB = XGBClassifier(random_state=7)
XGB_model =XGB.fit(X_train, y_train)
XGB_pred = XGB_model.predict(X_test_scaled)
XGB
```
